```python
# ============================================================
# NB 07. SurvSHAP(t) — 시간 의존적 설명 (survshap 0.4.2)
# Brain Tumor Survival Analysis — Explainable AI
#
# 임상 질문 3: 종양 등급의 영향이 시간에 따라 변하는가?
# 임상 질문 4: 치료 효과가 시간이 지남에 따라 지속되는가?
#
# STEP 1. 데이터 및 모델 불러오기
# STEP 2. SurvSHAP(t) 계산 (RSF + GBM)
# STEP 3. 임상 질문 3 — Grade 시간 의존적 효과
# STEP 4. 임상 질문 4 — 치료 효과 시간 의존성
# STEP 5. Schoenfeld 위반 변수 시각적 증명
# STEP 6. 고위험/저위험 환자 전체 비교
# STEP 7. 결과 저장
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
import warnings
import os
warnings.filterwarnings('ignore')

from survshap import SurvivalModelExplainer, PredictSurvSHAP
from statsmodels.nonparametric.smoothers_lowess import lowess

plt.rcParams.update({
    'font.family'      : 'AppleGothic',
    'axes.unicode_minus': False,
    'figure.dpi'       : 150,
    'savefig.dpi'      : 300,
    'savefig.bbox'     : 'tight',
    'axes.spines.top'  : False,
    'axes.spines.right': False,
})

OUTPUT_DIR = "./outputs/"
MODEL_DIR  = "./outputs/models/"
SEED       = 42
os.makedirs(OUTPUT_DIR, exist_ok=True)
os.makedirs(MODEL_DIR,  exist_ok=True)
np.random.seed(SEED)

# ============================================================
# STEP 1. 데이터 및 모델 불러오기
# ============================================================

print("="*55)
print("STEP 1. 데이터 및 모델 불러오기")
print("="*55)

ml_data    = joblib.load(MODEL_DIR + "ml_models.pkl")
local_xai  = joblib.load(MODEL_DIR + "local_xai.pkl")
schoenfeld = joblib.load(MODEL_DIR + "schoenfeld_result.pkl")

rsf_best     = ml_data['rsf_best']
gbm_best     = ml_data['gbm_best']
ML_FEATURES  = ml_data['ml_features']
X_train      = ml_data['X_train']
X_test       = ml_data['X_test']
y_train      = ml_data['y_train']
y_test       = ml_data['y_test']
times        = ml_data['times']

high_rep_idx  = local_xai['high_rep_idx']
low_rep_idx   = local_xai['low_rep_idx']
high_risk_idx = local_xai['high_risk_idx']
low_risk_idx  = local_xai['low_risk_idx']
risk_scores   = local_xai['risk_scores']
violated_vars = schoenfeld['violated_vars']

print(f"✓ 모델 불러오기 완료")
print(f"  Test 환자 수      : {len(X_test):,}명")
print(f"  평가 시간 구간    : {times.min():.0f}일 ~ {times.max():.0f}일")
print(f"  비례위험 위반 변수: {violated_vars}")

# ── 계산 대상 환자 선정 ───────────────────────────────────────
PATIENT_INDICES = {
    'high_risk': high_rep_idx,
    'low_risk' : low_rep_idx,
}

# Grade별 대표 환자 추가 (위험점수 중앙값 기준)
for grade_col, grade_label in [
    ('Grade_G4', 'G4'),
    ('Grade_G3', 'G3'),
    ('Grade_G2', 'G2'),
]:
    if grade_col not in ML_FEATURES:
        continue
    mask = X_test[grade_col].values == 1
    if mask.sum() == 0:
        continue
    grade_scores = risk_scores[mask]
    median_score = np.median(grade_scores)
    rel_idx      = np.argmin(np.abs(grade_scores - median_score))
    abs_idx      = np.where(mask)[0][rel_idx]
    PATIENT_INDICES[f'grade_{grade_label}'] = int(abs_idx)

# 치료별 대표 환자 선정 (실제 해당 치료를 받은 환자 중 위험점수 중앙값)
# → 각 치료 변수의 SHAP(t)는 해당 치료를 실제로 받은 환자에서만 신뢰 가능
TX_PATIENT_MAP = {}
print("\n[ 치료별 대표 환자 선정 ]")
for tx_col in ['Tx_Standard', 'Tx_Single', 'Tx_None']:
    if tx_col not in ML_FEATURES:
        print(f"  {tx_col}: 피처 없음 → 생략")
        continue
    mask = X_test[tx_col].values == 1
    n    = mask.sum()
    if n == 0:
        print(f"  {tx_col}: 해당 환자 없음 → 생략")
        continue
    tx_scores    = risk_scores[mask]
    median_score = np.median(tx_scores)
    rel_idx      = np.argmin(np.abs(tx_scores - median_score))
    abs_idx      = int(np.where(mask)[0][rel_idx])
    TX_PATIENT_MAP[tx_col] = abs_idx
    PATIENT_INDICES[f'tx_{tx_col}'] = abs_idx
    print(f"  {tx_col:12s}: n={n}, idx={abs_idx}, "
          f"위험점수={risk_scores[abs_idx]:.3f}")

print(f"\n계산 대상 환자 ({len(PATIENT_INDICES)}명):")
for k, v in PATIENT_INDICES.items():
    print(f"  {k:20s} → idx={v}, "
          f"위험점수={risk_scores[v]:.3f}")

# ============================================================
# STEP 2. SurvSHAP(t) 계산
# ============================================================
# survshap 0.4.2 구조:
#   result DataFrame 컬럼:
#     variable_str | variable_name | variable_value |
#     B | aggregated_change | t=T1 | t=T2 | ...
#   SHAP(t) 추출: B==0 행 필터 + 't = ' 컬럼 사용

print("\n" + "="*55)
print("STEP 2. SurvSHAP(t) 계산 (survshap 0.4.2)")
print("="*55)

def make_explainer(model, X_bg, y_tr):
    """
    SurvivalModelExplainer 생성
    predict_survival_function 래퍼 불필요 —
    survshap이 model.predict_survival_function 직접 호출
    """
    return SurvivalModelExplainer(
        model = model,
        data  = X_bg,
        y     = y_tr
    )


def compute_survshap_single(explainer, patient_data,
                             method='sampling', seed=42):
    """
    단일 환자 SurvSHAP(t) 계산
    survshap 0.4.2: PredictSurvSHAP().fit(explainer, obs)
    """
    result = PredictSurvSHAP(
        function_type      = 'sf',
        calculation_method = method,
        aggregation_method = 'integral',
        B                  = 25,
        random_state       = seed
    )
    result.fit(
        explainer       = explainer,
        new_observation = patient_data
    )
    return result


def compute_survshap_all(model, X_tr, y_tr, X_te,
                         patient_indices, model_name,
                         n_background=50):
    """
    전체 대상 환자 SurvSHAP(t) 계산
    treeshap 제거 → sampling으로 통일 (survshap 0.4.2 호환성)
    fallback: sampling → kernel
    """
    print(f"\n[ {model_name} SurvSHAP(t) ]")

    rng    = np.random.RandomState(SEED)
    bg_idx = rng.choice(len(X_tr), n_background, replace=False)
    X_bg   = X_tr.iloc[bg_idx]

    explainer = make_explainer(model, X_bg, y_tr)
    print(f"  ✓ Explainer 생성 완료 (배경 {n_background}개)")

    # treeshap 제거 → sampling으로 통일
    base_method = 'sampling'

    results = {}
    for label, idx in patient_indices.items():
        print(f"  [{label}] 계산 중... ", end='', flush=True)
        obs = X_te.iloc[[idx]]

        # 1차 시도: sampling
        try:
            res = compute_survshap_single(
                explainer, obs, method=base_method, seed=SEED
            )
            results[label] = res
            print("✓")
            continue
        except Exception as e1:
            print(f"\n    sampling 실패: {e1}")

        # 2차 시도: kernel
        try:
            res = compute_survshap_single(
                explainer, obs,
                method='kernel', seed=SEED
            )
            results[label] = res
            print(f"    ✓ kernel 재시도 성공")
        except Exception as e2:
            print(f"    [최종 실패] {e2}")
            results[label] = None

    success = sum(1 for v in results.values() if v is not None)
    print(f"\n  완료: {success}/{len(patient_indices)}명 성공")
    return results, explainer


# RSF SurvSHAP(t)
survshap_rsf, explainer_rsf = compute_survshap_all(
    rsf_best, X_train, y_train, X_test,
    PATIENT_INDICES, 'RSF'
)

# GBM SurvSHAP(t)
survshap_gbm, explainer_gbm = compute_survshap_all(
    gbm_best, X_train, y_train, X_test,
    PATIENT_INDICES, 'GBM'
)

# ============================================================
# 핵심 헬퍼 함수 — result에서 SHAP(t) 추출
# ============================================================

def extract_shap_t(survshap_result, feature_names):
    """
    survshap 0.4.2 result 구조에서 SHAP(t) 추출

    result 컬럼:
      variable_str | variable_name | variable_value |
      B | aggregated_change | t = T1 | t = T2 | ...

    반환:
      times_arr  : (n_times,) 시간 배열
      shap_dict  : {feature_name: shap_array} 딕셔너리
    """
    if survshap_result is None:
        return None, None

    try:
        df = survshap_result.result

        # B==0 행만 사용 (실제 SHAP 값)
        df_main = df[df['B'] == 0].copy()

        # 시간 컬럼 추출: 't = 숫자' 형태
        time_cols = [
            c for c in df_main.columns
            if str(c).startswith('t = ')
        ]
        if not time_cols:
            # 대안: 't=' 형태
            time_cols = [
                c for c in df_main.columns
                if str(c).startswith('t=')
            ]

        if not time_cols:
            print("  [경고] 시간 컬럼 없음")
            return None, None

        times_arr = np.array([
            float(str(c).replace('t = ', '').replace('t=', ''))
            for c in time_cols
        ])

        # 피처별 SHAP(t) 딕셔너리 구성
        shap_dict = {}
        for _, row in df_main.iterrows():
            feat = row['variable_name']
            shap_t = row[time_cols].values.astype(float)
            shap_dict[feat] = shap_t

        return times_arr, shap_dict

    except Exception as e:
        print(f"  [경고] extract 실패: {e}")
        return None, None


def get_feature_shap_t(survshap_result, feature_name, feature_names):
    """특정 피처의 SHAP(t) 배열 반환"""
    t_arr, shap_dict = extract_shap_t(
        survshap_result, feature_names
    )
    if t_arr is None:
        return None, None
    shap_t = shap_dict.get(feature_name)
    return t_arr, shap_t


# 추출 테스트
print("\n[ SHAP(t) 추출 테스트 ]")
for label in list(PATIENT_INDICES.keys())[:2]:
    res_rsf = survshap_rsf.get(label)
    t_arr, shap_dict = extract_shap_t(res_rsf, ML_FEATURES)
    if t_arr is not None:
        print(f"  RSF [{label}]: "
              f"시간 {len(t_arr)}개, "
              f"피처 {len(shap_dict)}개 추출 성공")
        print(f"  시간 범위: {t_arr.min():.0f} ~ "
              f"{t_arr.max():.0f}일")
    else:
        print(f"  RSF [{label}]: 추출 실패")

# ============================================================
# 공통 시각화 헬퍼
# ============================================================

CLINICAL_TIMES = [365, 730, 1095]

def add_clinical_vlines(ax, t_arr):
    """임상 시점 수직선 추가"""
    if t_arr is None:
        return
    for t_c, lbl in zip(CLINICAL_TIMES, ['1년', '2년', '3년']):
        if t_arr.min() <= t_c <= t_arr.max():
            ax.axvline(t_c, color='gray', linestyle='--',
                       linewidth=0.8, alpha=0.5)
            ymin, ymax = ax.get_ylim()
            ax.text(t_c, ymax * 0.92, lbl,
                    ha='center', fontsize=8, color='gray')


def add_lowess(ax, t_arr, shap_t, color, alpha=0.6):
    """LOWESS 추세선 추가"""
    try:
        if len(t_arr) > 5:
            smooth = lowess(shap_t, t_arr, frac=0.4)
            ax.plot(smooth[:, 0], smooth[:, 1],
                    color=color, linewidth=1.5,
                    linestyle='--', alpha=alpha)
    except Exception:
        pass


def plot_shap_t_line(ax, t_arr, shap_t, color,
                     label='', linewidth=2.5,
                     fill=True, show_lowess=True):
    """단일 SHAP(t) 곡선 그리기"""
    ax.plot(t_arr, shap_t, color=color,
            linewidth=linewidth, label=label)
    if fill:
        ax.fill_between(t_arr, shap_t, 0,
                        where=shap_t > 0,
                        alpha=0.15, color='#E05C5C')
        ax.fill_between(t_arr, shap_t, 0,
                        where=shap_t < 0,
                        alpha=0.15, color='#4C9BE8')
    if show_lowess:
        add_lowess(ax, t_arr, shap_t, color)
    ax.axhline(0, color='black', linestyle=':',
               linewidth=1, alpha=0.5)
    add_clinical_vlines(ax, t_arr)


# ============================================================
# STEP 3. 임상 질문 3 — Grade 시간 의존적 효과
# ============================================================

print("\n" + "="*55)
print("STEP 3. 임상 질문 3 — Grade 시간 의존적 효과")
print("="*55)
print("Grade G4/G3/G2 환자에서 Grade 변수의 SHAP(t) 변화")
print("곡선이 평탄하지 않으면 → 시간 의존적 효과 존재")

GRADE_MAP = {
    'Grade_G4': ('grade_G4', '#E05C5C'),
    'Grade_G3': ('grade_G3', '#F0A500'),
    'Grade_G2': ('grade_G2', '#4C9BE8'),
}

for model_name, survshap_dict in [
    ('RSF', survshap_rsf),
    ('GBM', survshap_gbm)
]:
    fig, axes = plt.subplots(1, 3, figsize=(16, 5))
    fig.suptitle(
        f"{model_name} — Grade별 SurvSHAP(t) 곡선\n"
        "임상 질문 3: Grade 효과가 시간에 따라 변하는가?",
        fontsize=13, fontweight='bold'
    )

    for ax, (grade_feat, (patient_key, color)) in zip(
        axes, GRADE_MAP.items()
    ):
        result = survshap_dict.get(patient_key)

        if result is None:
            ax.text(0.5, 0.5, '계산 실패\n(데이터 부족)',
                    ha='center', va='center',
                    transform=ax.transAxes, fontsize=11,
                    color='#888888')
            ax.set_title(f"{grade_feat} 환자",
                         fontsize=11, fontweight='bold')
            continue

        # 모든 Grade 변수의 SHAP(t) 표시
        t_ref = None
        for feat, (_, c) in GRADE_MAP.items():
            t_arr, shap_t = get_feature_shap_t(
                result, feat, ML_FEATURES
            )
            if t_arr is None:
                continue
            t_ref = t_arr
            is_main = (feat == grade_feat)
            plot_shap_t_line(
                ax, t_arr, shap_t,
                color   = c,
                label   = feat,
                linewidth = 2.5 if is_main else 1.2,
                fill    = is_main,
                show_lowess = is_main
            )

        ax.set_title(
            f"{grade_feat} 환자",
            fontsize=11, fontweight='bold', color=color
        )
        ax.set_xlabel("Time (Days)", fontsize=10)
        ax.set_ylabel("SHAP(t)", fontsize=10)
        ax.legend(fontsize=8)

    plt.tight_layout()
    fname = f"fig_survshap_grade_{model_name.lower()}.png"
    plt.savefig(OUTPUT_DIR + fname)
    plt.show()
    print(f"✓ {fname} 저장 완료")

# ============================================================
# STEP 4. 임상 질문 4 — 치료 효과 시간 의존성
# ============================================================

print("\n" + "="*55)
print("STEP 4. 임상 질문 4 — 치료 효과 시간 의존성")
print("="*55)
print("Tx_Single / Tx_None의 SHAP(t) 곡선")
print("시간에 따라 감소하면 → 치료 효과 약화")

# 치료별 색상 정의
TX_FEATURES = {
    'Tx_Standard': '#2ECC71',  # 초록 — 표준 치료 (기준)
    'Tx_Single'  : '#F0A500',  # 주황 — 단일 치료
    'Tx_None'    : '#E05C5C',  # 빨강 — 무치료
}

# 치료별 대표 환자가 없으면 STEP 4 경고 후 진행
if not TX_PATIENT_MAP:
    print("  [경고] 치료별 대표 환자 선정 실패 → STEP 4 생략")
else:
    for model_name, survshap_dict in [
        ('RSF', survshap_rsf),
        ('GBM', survshap_gbm)
    ]:
        # subplot: 치료별 대표 환자 1명씩 — 각 subplot에 해당 치료 SHAP(t) 표시
        # X축: 치료 종류 (Tx_Standard / Tx_Single / Tx_None)
        # 각 subplot: 해당 치료를 실제로 받은 대표 환자의 모든 치료 변수 SHAP(t) 비교
        n_tx  = len(TX_PATIENT_MAP)
        fig, axes = plt.subplots(1, n_tx, figsize=(n_tx * 6, 5))
        if n_tx == 1:
            axes = [axes]

        fig.suptitle(
            f"{model_name} — 치료 변수 SurvSHAP(t) 곡선\n"
            "임상 질문 4: 치료 효과가 시간에 따라 지속되는가? "
            "(각 치료를 실제로 받은 대표 환자)",
            fontsize=13, fontweight='bold'
        )

        for ax, (tx_feat, pat_idx) in zip(axes, TX_PATIENT_MAP.items()):
            result = survshap_dict.get(f'tx_{tx_feat}')
            pat_label = (
                f"{tx_feat} 환자\n"
                f"(idx={pat_idx}, "
                f"위험점수={risk_scores[pat_idx]:.3f})"
            )

            if result is None:
                ax.text(0.5, 0.5, '계산 실패',
                        ha='center', va='center',
                        transform=ax.transAxes, fontsize=11,
                        color='#888888')
                ax.set_title(pat_label, fontsize=10,
                             fontweight='bold')
                continue

            # 모든 치료 변수 SHAP(t) 겹쳐 표시
            for i, (feat, color) in enumerate(TX_FEATURES.items()):
                t_arr, shap_t = get_feature_shap_t(
                    result, feat, ML_FEATURES
                )
                if t_arr is None:
                    continue

                is_main = (feat == tx_feat)
                plot_shap_t_line(
                    ax, t_arr, shap_t,
                    color     = color,
                    label     = feat,
                    linewidth = 2.8 if is_main else 1.5,
                    fill      = is_main,
                    show_lowess = True
                )

                # 증가/감소 추세 표시
                if len(t_arr) > 3:
                    slope = np.polyfit(t_arr, shap_t, 1)[0]
                    trend = '↑ 효과 증가' if slope > 0 \
                            else '↓ 효과 감소'
                    ax.text(
                        0.98, 0.02 + i * 0.1,
                        f"{feat}: {trend}",
                        transform=ax.transAxes,
                        ha='right', va='bottom',
                        fontsize=8, color=color
                    )

            # 해당 치료 변수 강조 표시
            ax.set_title(pat_label, fontsize=10,
                         fontweight='bold',
                         color=TX_FEATURES.get(tx_feat, 'black'))
            ax.set_xlabel("Time (Days)", fontsize=10)
            ax.set_ylabel("SHAP(t)", fontsize=10)
            ax.legend(fontsize=9)

        plt.tight_layout()
        fname = f"fig_survshap_tx_{model_name.lower()}.png"
        plt.savefig(OUTPUT_DIR + fname)
        plt.show()
        print(f"✓ {fname} 저장 완료")

# ============================================================
# STEP 5. Schoenfeld 위반 변수 시각적 증명
# ============================================================
# 논문 핵심 차별점:
# Cox에서 가정 위반된 변수가
# SurvSHAP(t)에서 시간에 따라 변하는 곡선을 보이면
# → ML+XAI가 Cox 한계를 극복한 것을 시각적 증명

print("\n" + "="*55)
print("STEP 5. Schoenfeld 위반 변수 시각적 증명")
print("="*55)
print(f"위반 변수: {violated_vars}")
print("SurvSHAP(t) 곡선이 평탄하지 않으면")
print("→ 비례위험 가정 위반을 시각적으로 증명")

if not violated_vars:
    print("위반 변수 없음 → STEP 5 생략")
else:
    for model_name, survshap_dict in [
        ('RSF', survshap_rsf),
        ('GBM', survshap_gbm)
    ]:
        n_vars = len(violated_vars)
        fig, axes = plt.subplots(
            1, n_vars,
            figsize=(n_vars * 5.5, 5)
        )
        if n_vars == 1:
            axes = [axes]

        fig.suptitle(
            f"{model_name} — Schoenfeld 위반 변수 SurvSHAP(t)\n"
            "Cox 비례위험 가정 위반의 시각적 증명",
            fontsize=13, fontweight='bold'
        )

        for ax, var in zip(axes, violated_vars):
            # 고위험/저위험 환자 비교
            plotted = False
            for patient_key, color, label in [
                ('high_risk', '#E05C5C', '고위험'),
                ('low_risk',  '#4C9BE8', '저위험'),
            ]:
                result = survshap_dict.get(patient_key)
                t_arr, shap_t = get_feature_shap_t(
                    result, var, ML_FEATURES
                )
                if t_arr is None or shap_t is None:
                    continue

                plot_shap_t_line(
                    ax, t_arr, shap_t,
                    color=color, label=label,
                    linewidth=2.5, fill=False,
                    show_lowess=True
                )
                plotted = True

            if not plotted:
                ax.text(0.5, 0.5, '계산 실패',
                        ha='center', va='center',
                        transform=ax.transAxes,
                        fontsize=11, color='#888888')

            # 평탄도 검정 (선형회귀 기울기)
            result_h = survshap_dict.get('high_risk')
            t_h, s_h = get_feature_shap_t(
                result_h, var, ML_FEATURES
            )
            if t_h is not None and s_h is not None:
                slope = np.polyfit(t_h, s_h, 1)[0]
                flatness = '비평탄 (시간 의존성 ✓)' \
                           if abs(slope) > 1e-5 \
                           else '평탄 (시간 불변)'
                ax.text(
                    0.02, 0.98, f"기울기: {slope:.2e}\n{flatness}",
                    transform=ax.transAxes,
                    ha='left', va='top', fontsize=8,
                    bbox=dict(boxstyle='round,pad=0.3',
                              facecolor='white', alpha=0.8)
                )

            ax.set_title(
                f"{var}\n(Schoenfeld p<0.05 — 비례위험 위반)",
                fontsize=10, fontweight='bold',
                color='#E05C5C'
            )
            ax.set_xlabel("Time (Days)", fontsize=9)
            ax.set_ylabel("SHAP(t)", fontsize=9)
            ax.legend(fontsize=9)

        plt.tight_layout()
        fname = f"fig_survshap_violated_{model_name.lower()}.png"
        plt.savefig(OUTPUT_DIR + fname)
        plt.show()
        print(f"✓ {fname} 저장 완료")

# ============================================================
# STEP 6. 고위험/저위험 환자 상위 변수 전체 비교
# ============================================================

print("\n" + "="*55)
print("STEP 6. 고위험/저위험 환자 전체 비교")
print("="*55)

global_xai_data = joblib.load(MODEL_DIR + "global_xai.pkl")

for model_name, survshap_dict, df_shap in [
    ('RSF', survshap_rsf, global_xai_data['df_shap_rsf']),
    ('GBM', survshap_gbm, global_xai_data['df_shap_gbm']),
]:
    top_n        = 6
    top_features = df_shap['Feature'].head(top_n).tolist()

    fig, axes = plt.subplots(
        2, top_n,
        figsize=(top_n * 3.5, 8)
    )
    fig.suptitle(
        f"{model_name} — 상위 {top_n}개 변수 SurvSHAP(t)\n"
        "고위험 vs 저위험 환자 비교",
        fontsize=13, fontweight='bold'
    )

    for col_i, feat in enumerate(top_features):
        for row_i, (patient_key, color, label) in enumerate([
            ('high_risk', '#E05C5C', '고위험'),
            ('low_risk',  '#4C9BE8', '저위험'),
        ]):
            ax     = axes[row_i, col_i]
            result = survshap_dict.get(patient_key)
            t_arr, shap_t = get_feature_shap_t(
                result, feat, ML_FEATURES
            )

            if t_arr is None or shap_t is None:
                ax.text(0.5, 0.5, 'N/A',
                        ha='center', va='center',
                        transform=ax.transAxes,
                        color='#888888')
            else:
                plot_shap_t_line(
                    ax, t_arr, shap_t,
                    color=color, linewidth=2,
                    fill=True, show_lowess=False
                )

            if row_i == 0:
                ax.set_title(feat, fontsize=9,
                             fontweight='bold')
            ax.set_ylabel(f"{label}\nSHAP(t)",
                          fontsize=8, color=color)
            if row_i == 1:
                ax.set_xlabel("Time (Days)", fontsize=8)
            ax.tick_params(labelsize=7)

    plt.tight_layout()
    fname = f"fig_survshap_full_{model_name.lower()}.png"
    plt.savefig(OUTPUT_DIR + fname)
    plt.show()
    print(f"✓ {fname} 저장 완료")

# ============================================================
# STEP 7. 결과 저장
# ============================================================

print("\n" + "="*55)
print("STEP 7. 결과 저장")
print("="*55)

joblib.dump({
    'survshap_rsf'   : survshap_rsf,
    'survshap_gbm'   : survshap_gbm,
    'patient_indices': PATIENT_INDICES,
    'tx_patient_map' : TX_PATIENT_MAP,
    'violated_vars'  : violated_vars,
    'ml_features'    : ML_FEATURES,
    'times'          : times,
}, MODEL_DIR + "survshap_result.pkl")

print("✓ survshap_result.pkl 저장 완료")

# ── 성공 여부 최종 확인 ───────────────────────────────────────
print(f"\n{'='*55}")
print("NB 07 완료 — SurvSHAP(t) 최종 요약")
print(f"{'='*55}")

for model_name, survshap_dict in [
    ('RSF', survshap_rsf),
    ('GBM', survshap_gbm)
]:
    success = sum(
        1 for v in survshap_dict.values() if v is not None
    )
    total   = len(survshap_dict)
    print(f"\n  {model_name}: {success}/{total}명 성공")
    for label, result in survshap_dict.items():
        status = '✓' if result is not None else '✗'
        print(f"    {status} {label}")

print(f"\n[ 임상 질문 3: Grade 시간 의존적 효과 ]")
print(f"  → fig_survshap_grade_rsf/gbm.png")
print(f"  → 곡선 기울기 ≠ 0 이면 시간 의존성 존재")

print(f"\n[ 임상 질문 4: 치료 효과 시간 지속성 ]")
print(f"  → fig_survshap_tx_rsf/gbm.png")
print(f"  → SHAP(t) 감소 추세면 치료 효과 약화")

print(f"\n[ Schoenfeld 위반 변수 시각적 증명 ]")
print(f"  위반 변수: {violated_vars}")
print(f"  → fig_survshap_violated_rsf/gbm.png")
print(f"  → Cox 가정 위반을 SurvSHAP(t)로 시각적 증명")
print(f"{'='*55}")
print("\n→ 다음: 논문 NB08 Figure 최종 Export")
```
