```python
import pandas as pd

# TSV 파일 읽기
df = pd.read_csv("clinical.tsv", sep="\t")

# CSV 파일로 저장
df.to_csv("clinical.csv", index=False)
```

```python
print(df.shape)
print(df.columns)
print(df.head())
```

```python
import pandas as pd

df = df.replace(["'--", "--"], pd.NA)
```

```python
missing_ratio = df.isna().mean().sort_values(ascending=False)
missing_ratio.head(20)
```

```python
# 80% 이상 결측 컬럼
drop_cols = missing_ratio[missing_ratio >= 0.8].index

df_reduced = df.drop(columns=drop_cols)
```

```python
df_reduced.columns
```

```python
print("삭제된 컬럼 수:", len(drop_cols))
for col in drop_cols:
    print(col)
```

```python
missing_count = df_reduced.isna().sum().sort_values(ascending=False)

missing_count
```
